# Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\Data Analyst\Python\Pandas\LOTR.csv")

df1
```

```python
df2 = pd.read_csv(r"D:\Data Analyst\Python\Pandas\LOTR 2.csv")

df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix ='_left', rsuffix ='_right')
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix ='_left', rsuffix ='_right', how = 'outer')
df3
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
