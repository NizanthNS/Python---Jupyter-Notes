# If - Elif - Else

```python
if 25 > 10 :
    print("It's True!")
```

```python
if 25 < 10 :
    print("It's True!")
else:
    print("It's not True")
```

```python
if 25 < 10 :
    print("It's True!")
elif 25 < 30 :
    print("It's True!")
else:
    print("It's not True")
```

```python
if 25 < 10 :
    print("It's True!")
elif 25 < 20 :
    print("Elif 1 It's True!")
elif 25 < 21 :
    print("Elif 2 It's True!")
elif 25 < 30 :
    print("Elif 3 It's True!")
elif 25 < 40 :
    print("Elif 4 It's True!")
else:
    print("It's not True")
```

```python
if (25 < 10) or (1 < 3) :
    print("It's True!")
elif 25 < 20 :
    print("Elif 1 It's True!")
elif 25 < 21 :
    print("Elif 2 It's True!")
elif 25 < 30 :
    print("Elif 3 It's True!")
elif 25 < 40 :
    print("Elif 4 It's True!")
else:
    print("It's not True")
```

```python
print("It's true") if 10 > 30 else print("It's not true")
```

```python
if (25 < 10) or (1 < 3) :
    print("It's True!")
    if 10 > 5:
        print("This nested if statement is True")
elif 25 < 20 :
    print("Elif 1 It's True!")
elif 25 < 21 :
    print("Elif 2 It's True!")
elif 25 < 30 :
    print("Elif 3 It's True!")
elif 25 < 40 :
    print("Elif 4 It's True!")
else:
    print("It's not True")
```
