# While Loops

```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer less than 5')
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        continue
    number = number + 1
else:
    print('No longer less than 5') # Cause infinite loop
```

```python
number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)    
else:
    print('No longer less than 5')
```
