# For Loops

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print(number)
```

```python
for number in integers:
    print('LOL')
```

```python
integers = [1,2,3,4,5]
for Jelly in integers:
    print(Jelly + Jelly)
```

```python
dict_food = {'Name' : 'Niz', 'Daily purchase' : 2, 'Favorite fried chicken' : ['Bigshot', 'Alfam']}
```

```python
for food in dict_food.values():
    print(food)
```

```python
for key, value in dict_food.items():
    print(key, '-', value)
```

# Nested For Loops

```python
flavours = ['Vanilla', 'Chocolate', 'Cookie Crackers']
toppings = ['Hot Fudge', 'Oreo', 'Marshmellows']
```

```python
for one in flavours:
    for two in toppings:
        print(one, 'topped with', two)
```
