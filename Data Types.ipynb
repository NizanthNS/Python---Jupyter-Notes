```python
# Sets - Unique

Theme = {1,2,3}
```

```python
type(Theme)
```

```python
print(Theme)
```

```python
Theme_types = {1,2,3,1,2,3,4,5,6,4,6,5}

print(Theme_types)
```

```python
Tune_types = {1,2,3,12,34,5,64,6,5,8}
```

```python
print(Theme_types | Tune_types)
```

```python
print(Theme_types & Tune_types)
```

```python
print(Theme_types - Tune_types)
```

```python
print(Tune_types - Theme_types)
```

```python
print(Theme_types ^ Tune_types)
```
