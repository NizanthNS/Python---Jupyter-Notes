```python
def number_square(number):
    print(number*2)
```

```python
def number_square_1(number, power): # orbitary argument
    print(number**power)
```

```python
number_square_1(5, 4)
```

```python
args_tuple = (5,6,1,2,8)
def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(*args_tuple)
```

```python
def number_key(**number):  # Orbitary keyword argument
    print('My number is: ' + number['integer'] + ' My other number is: ' + number['integer2'])
```
